# **EXPLORACIÓN DE DATOS**
Este es el notebook para realizar la exploración profunda de los datos. 

Primero, es necesario importar los datos limpios.

```python
from pandas import read_csv
from os import path, getcwd
directorio_padre = path.dirname(getcwd())
datos = read_csv(directorio_padre + '\\datos\\datos_limpios.csv')
print(datos.head())
```

### Estadísticas descriptivas.

```python
# Analizar las variables explicativas
print(datos['incentive'].unique())
print(datos['no_of_workers_redondeado'].unique())

# Analizar vaar. de respuesta
print(datos['actual_productivity'].unique())

# Estadísticas decr.
print(datos.iloc[:,[7,8,10]].describe())
```

Se pueden realizar las siguientes observaciones de la variable de respuesta:
- La media y la mediana son muy similares. Esto podría sugerir que la distribución de los datos es simétrica o que no hay valores atípicos significativos que estén sesgando la media. La mediana es ligeramente mayor que la media, lo cual podría indicar una ligera asimetría negativa en la distribución, es decir, hay algunos valores relativamente bajos que están afectando ligeramente la media hacia abajo. Se podría suponer de antemano que la media y la mediana pueden proporcionar una estimación confiable de la tendencia central de los datos.
- La razón desviación estándar/media es de 0.237, es decir, el coeficiente de variación es del 23.7%. Esto puede sugerir una variabilidad relativamente baja en comparación con la media, y que los datos tienden a agruparse alrededor de la media.
- El rango de valores es de aproximadamente 0.886, lo que indica una diferencia considerable entre el valor mínimo y máximo de productividad observado en el equipo.
- El primer cuartil (25%) es 0.650307 y el tercer cuartil (75%) es 0.850253. Esto sugiere que el 50% de los datos están comprendidos entre estos dos valores. El rango intercuartil es de aproximadamente 0.2, lo que indica la dispersión de los datos en esta parte central de la distribución.

Por otro lado, se pueden realizar las siguientes observaciones de las variable de explicativas:

1. **Incentivo Económico**:
   - El valor medio del incentivo económico es de aproximadamente 38.21. Esto indica que, en promedio, se ofrece un incentivo económico a los trabajadores.
   - La desviación estándar del incentivo económico es alta, con un valor de alrededor de 160.18. Esto sugiere una gran variabilidad en los montos de incentivo económico ofrecidos.
   - El rango de valores observado para el incentivo económico va desde 0 hasta 3600. Esto indica una amplia variación en los montos de incentivo ofrecidos, desde montos muy bajos hasta montos muy altos.
   - El primer cuartil (25%) y el tercer cuartil (75%) son 0 y 50, respectivamente. Esto indica que el 50% de los datos se encuentra entre estos dos valores, lo que sugiere una dispersión considerable en los montos de incentivo económico ofrecidos.

2. **Número de Trabajadores**:
   - El número medio de trabajadores es de aproximadamente 34.55. Esto sugiere que, en promedio, hay alrededor de 34 o 35 trabajadores en cada equipo.
   - La desviación estándar del número de trabajadores es de alrededor de 22.15, lo que indica una moderada variabilidad en la cantidad de trabajadores en cada equipo.
   - El rango de valores observado para el número de trabajadores va desde 2 hasta 89. Esto indica una amplia variación en el tamaño de los equipos de trabajo.
   - El primer cuartil (25%) y el tercer cuartil (75%) son 9 y 57, respectivamente. Esto indica que el 50% de los datos se encuentra entre estos dos valores, lo que sugiere una dispersión considerable en el tamaño de los equipos de trabajo.

### Boxplots.

```python
import matplotlib.pyplot as plt

datos.boxplot(column='incentive')
plt.tight_layout()
plt.show()

datos.boxplot(column='no_of_workers_redondeado')
plt.tight_layout()
plt.show()

datos.boxplot(column='actual_productivity')
plt.tight_layout()
plt.show()
```

### Pairplot.

```python
from seaborn import pairplot
%matplotlib inline
pairplot(datos.iloc[:,[7,8,10]])
```

Se pueden evidenciar 10 datos atípicos en la variable de respuesta que afectan mucho la distribución de los datos. Por ende, se procederá a quitarlos.

```python
# Filtrar los valores atípicos
datos_corregidos = datos[(datos['incentive'] < 960)]

# Ver boxplot sin los datos atípicos
datos_corregidos.boxplot(column='incentive')
plt.tight_layout()
plt.show()

# Ver histograma sin los datos atípicos
%matplotlib inline
pairplot(datos_corregidos.iloc[:,[7]])
```

### Matriz de correlación.

```python
from seaborn import heatmap
corr = datos_corregidos.iloc[:,[7,8,10]].corr()
heatmap(corr, cmap="Blues", annot=True)
```

Se pueden realizar las siguientes observaciones:
   
1. La correlación entre la variable de respuesta (productividad) y la variable predictora incentivo económico es 0.31. Esto sugiere una correlación positiva débil entre el incentivo económico y la productividad. Aunque la correlación es positiva, la magnitud (0.31) indica que la relación no es muy fuerte.

2. La correlación entre la variable de respuesta (productividad) y la variable predictora número de trabajadores del equipo es 0.69. Esto sugiere una correlación positiva más fuerte entre el número de trabajadores del equipo y la productividad en comparación con el incentivo económico. Una correlación más alta indica una relación más fuerte entre estas variables.

La correlación más alta entre la productividad y el número de trabajadores del equipo sugiere que esta variable puede ser un predictor más fuerte de la productividad en comparación con el incentivo económico. La correlación positiva entre el incentivo económico y la productividad indica que un mayor incentivo económico se asocia con una mayor productividad, pero la correlación es relativamente débil.

Sin embargo, es importante recordar que la correlación no implica causalidad, y se necesitarían análisis adicionales, como una regresión lineal, para comprender completamente la relación entre estas variables.

```python
pairplot(datos_corregidos, x_vars=['no_of_workers_redondeado','actual_productivity'],
          y_vars='incentive',height=7, aspect=0.7, kind='reg')
```
